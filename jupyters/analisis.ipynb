# Análisis de variables numéricas
- `store_number`

- `county_number`
     - Es la id de los condados, no es necesario analizarlo
- `category_number`
     - Es la id de los condados, no es necesario analizarlo
- `vendor_number`
     - Es la id de los vendedores, no es necesario analizarlo
- `item_number`
     - Es la id de los productos, no es necesario analizarlo
- `pack` (nº de botellas por pack)
     - tiene dos outliers de packs que tienen unas 25 y 50 botellas
     - los que más destacan son los packs de 12 seguido de pack de 6
- `bottle_volume(ml)`
     - outliers en 1750 por arriba y por debajo pero pueden tener sentido
     - la mayoria de los datos se encuentra en las botellas de 750 a 1750 ml
- `state_bottle_cost` (impuesto) y `state_bottle_retail`
     - forma bastante parecida, parece que ambas categorías estan estrechamente relacionadas.
     - la distribucion se centra en los valores mas bajos y hay outliers por arriba
- `bottles_sold`
     - hay valores muy altos como outliers, es posible que no sean errores, ya que el valor maximo es 300(aprox)
     - la distribucion se centra en los valores mas bajos
- `sale(dollars)`
     - centrado a la izda (valores bajos), muchos outliers bastante altos
     - habria que ver si tiene sentido con el numero de botellas vendidas y el precio.
          - si provienen de las botellas mas caras o de los pedidos mas grandes
- `volume_sold(gallons)` parecido al caso anterior, tenemos otra variable creada en ml
- `latitude` y `longitude`
     - no es necesasrio el análisis
